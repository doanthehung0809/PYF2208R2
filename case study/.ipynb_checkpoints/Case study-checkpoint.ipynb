```python
from tkinter import *
from tkinter import ttk
from tkinter import messagebox
class teacher:
    def __init__(self, MaGV, TenGV, Sex, Khoa, Sdt, Email, DiaChi):
        self.MaGV = MaGV
        self.TenGV = TenGV
        self.Sex = Sex
        self.Khoa = Khoa
        self.Sdt = Sdt
        self.Email = Email
        self.DiaChi = DiaChi
list_teacher = []
def add_info():
    list_teacher.append(teacher(entry_MaGV.get(), entry_TenGV.get(), combo_Sex.get(), entry_Khoa.get(), entry_Sdt.get(), entry_Email.get(), entry_DiaChi.get()))
    Teacher_Table.insert(parent='',index='end',values=(entry_MaGV.get(), entry_TenGV.get(), combo_Sex.get(), entry_Khoa.get(), entry_Sdt.get(), entry_Email.get(), entry_DiaChi.get()))

def reset_info():
    entry_MaGV.delete(0, len(entry_MaGV.get()))
    entry_TenGV.delete(0, len(entry_TenGV.get()))
    entry_Khoa.delete(0, len(entry_Khoa.get()))
    entry_Sdt.delete(0, len(entry_Sdt.get()))
    entry_Email.delete(0, len(entry_Email.get()))
    entry_DiaChi.delete(0, len(entry_DiaChi.get()))
def delete_tree():
    selected_item = Teacher_Table.selection()[0]
    Teacher_Table.delete(selected_item)
def clear_all():
    for item in Teacher_Table.get_children():
        Teacher_Table.delete(item)
def delete_info():
    delete_MaGV = entry_MaGV.get()
    exist = False
    for item in range(len(list_teacher)):
        if list_teacher[item].MaGV == entry_MaGV.get():
            exist = True
            del list_teacher[item]
            clear_all()
            for item in list_teacher:
                Teacher_Table.insert(parent='',index='end',values=(item.MaGV, item.TenGV, item.Sex, item.Khoa, item.Sdt, item.Email, item.DiaChi))
            messagebox.showinfo("Xoá Thông Tin Giáo Viên", f"Giáo viên mã số: {delete_MaGV} đã được xoá")
            break
    if exist == False:
        messagebox.showinfo("Xoá Thông Tin Giáo Viên", f"Giáo viên mã số: {delete_MaGV} khong ton tai")
def edit_info():
    check = entry_MaGV.get()
    exist = False
    for item in range(len(list_teacher)):
        if list_teacher[item].MaGV == check:
            list_teacher[item].MaGV = entry_MaGV.get()
            list_teacher[item].TenGV = entry_TenGV.get()
            list_teacher[item].Sex = combo_Sex.get()
            list_teacher[item].Khoa = entry_Khoa.get()
            list_teacher[item].Sdt = entry_Sdt.get()
            list_teacher[item].Email = entry_Email.get()
            list_teacher[item].DiaChi = entry_DiaChi.get()
            clear_all()
            for item in list_teacher:
                Teacher_Table.insert(parent='',index='end',values=(item.MaGV, item.TenGV, item.Sex, item.Khoa, item.Sdt, item.Email, item.DiaChi))
            messagebox.showinfo("Sửa Thông Tin Giáo Viên", f"Dữ liệu giáo viên mã số: {list_teacher[item].MaGV}, tên: {list_teacher[item].TenGV} đã được sửa")
            exist = True
            break
    if exist == False:
        messagebox.showinfo("Sửa Thông Tin Giáo Viên", f"Giáo viên mã số: {check} không tìm được trong danh sách")
def search_info():
    search = text_Search.get()
    exist = False
    if combo_Search.get() == "Tên GV":
        for item in list_teacher:
            if item.TenGV == search:
                exist = True
                messagebox.showinfo("Tìm Kiếm", f"Giáo viên tên: {search} có tồn tại trong danh sách. Mã GV: {item.MaGV}, Tên GV: {item.TenGV}, Giới tính: {item.Sex}, Khoa: {item.Khoa}, SĐT: {item.Sdt}, Email: {item.Email}")
                break
    elif combo_Search.get() == "Mã GV":
        for item in list_teacher:
            if item.MaGV == search:
                exist = True
                messagebox.showinfo("Tìm Kiếm", f"Mã giáo viên: {search} có tồn tại trong danh sách. Mã GV: {item.MaGV}, Tên GV: {item.TenGV}, Giới tính: {item.Sex}, Khoa: {item.Khoa}, SĐT: {item.Sdt}, Email: {item.Email}")
                break
    elif combo_Search.get() == "Khoa":
        for item in list_teacher:
            if item.Khoa == search:
                exist = True
                messagebox.showinfo("Tìm Kiếm", f"Khoa: {search} có tồn tại trong danh sách. Mã GV: {item.MaGV}, Tên GV: {item.TenGV}, Giới tính: {item.Sex}, Khoa: {item.Khoa}, SĐT: {item.Sdt}, Email: {item.Email}")
                break
    if exist == False:
        messagebox.showinfo("Tìm Kiếm", f"{search} không tồn tại trong danh sách")

root = Tk()
root.title("Teacher Managerment System")
root.geometry('1000x700+0+0')
title = Label(root, text="Teacher Managerment System", font=("arial",40,"bold"), fg="white", bg="navy")
title.pack(side=TOP, fill=X)

Manager_Frame = Frame(root, bd=4, relief=RIDGE, bg="red")
Manager_Frame.place(x=10, y=70, width=400, height=550)

m_title = Label(Manager_Frame, text="MANAGER TEACHER", font=("arial",16,"bold"), bg="red", fg="white")
m_title.grid(row=0, columnspan=2, pady=10, padx=80)
        
lbl_MaGV = Label(Manager_Frame, text="Mã GV", font=("arial",12,"bold"), bg="red", fg="white")
lbl_MaGV.grid(row=1, column=0, pady=10, padx=20, sticky="w")     
entry_MaGV = Entry(Manager_Frame, font=("arial",12), bd=5, relief=GROOVE)
entry_MaGV.grid(row=1, column=1, pady=10, padx=20, sticky="w")
        
lbl_TenGV = Label(Manager_Frame, text="Tên GV", font=("arial",12,"bold"), bg="red", fg="white")
lbl_TenGV.grid(row=2, column=0, pady=10, padx=20, sticky="w")
entry_TenGV = Entry(Manager_Frame, font=("arial",12), bd=5, relief=GROOVE)
entry_TenGV.grid(row=2, column=1, pady=10, padx=20, sticky="w")
        
lbl_Sex = Label(Manager_Frame, text="Giới tính", font=("arial",12,"bold"), bg="red", fg="white")
lbl_Sex.grid(row=3, column=0, pady=10, padx=20, sticky="w")
combo_Sex = ttk.Combobox(Manager_Frame, font=("arial",11), state="readonly")
combo_Sex['values']=("Nam","Nữ","Khác")
combo_Sex.grid(row=3, column=1, ipady=1.5, ipadx=1)
        
lbl_Khoa = Label(Manager_Frame, text="Khoa", font=("arial",12,"bold"), bg="red", fg="white")
lbl_Khoa.grid(row=4, column=0, pady=10, padx=20, sticky="w")
entry_Khoa = Entry(Manager_Frame, font=("arial",12), bd=5, relief=GROOVE)
entry_Khoa.grid(row=4, column=1, pady=10, padx=20, sticky="w")
        
lbl_Sdt = Label(Manager_Frame, text="Số điện thoại", font=("arial",12,"bold"), bg="red", fg="white")
lbl_Sdt.grid(row=5, column=0, pady=10, padx=20, sticky="w")
entry_Sdt = Entry(Manager_Frame, font=("arial",12), bd=5, relief=GROOVE)
entry_Sdt.grid(row=5, column=1, pady=10, padx=20, sticky="w")

lbl_Email = Label(Manager_Frame, text="Email", font=("arial",12,"bold"), bg="red", fg="white")
lbl_Email.grid(row=6, column=0, pady=10, padx=20, sticky="w")
entry_Email = Entry(Manager_Frame, font=("arial",12), bd=5, relief=GROOVE)
entry_Email.grid(row=6, column=1, pady=10, padx=20, sticky="w")
        
lbl_DiaChi = Label(Manager_Frame, text="Địa Chỉ", font=("arial",12,"bold"), bg="red", fg="white")
lbl_DiaChi.grid(row=7, column=0, pady=10, padx=20, sticky="w")
entry_DiaChi = Entry(Manager_Frame, font=("arial",12), bd=5, relief=GROOVE)
entry_DiaChi.grid(row=7, column=1, pady=10, padx=20, sticky="w")
        
#Button Frame
Btn_Frame = Frame(root, bd=4, relief=RIDGE, bg="white")
Btn_Frame.place(x=10, y=530, width=400)
        
Btn_Add = Button(Btn_Frame, text="Thêm", width=10, command=add_info)
Btn_Add.grid(row=0, column=0, pady=10, padx=10)
Btn_Reset = Button(Btn_Frame, text="Reset", width=10, command=reset_info)
Btn_Reset.grid(row=0, column=1, pady=10, padx=10)
Btn_Edit = Button(Btn_Frame, text="Sửa", width=10, command=edit_info)
Btn_Edit.grid(row=0, column=2, pady=10, padx=10)
Btn_Delete = Button(Btn_Frame, text="Xoá", width=10, command=delete_info)
Btn_Delete.grid(row=0, column=3, pady=10, padx=10)
        
# Manager detail       
Detail_Frame = Frame(root, bd=4, relief=RIDGE, bg="red")
Detail_Frame.place(x=420, y=70, width=850, height=550)
        
lbl_Search = Label(Detail_Frame, text="Search By", font=("arial",12,"bold"), bg="red", fg="white")
lbl_Search.grid(row=0, column=0, pady=10, padx=20, sticky="w")
        
combo_Search = ttk.Combobox(Detail_Frame, width=10, font=("arial",11), state="readonly")
combo_Search['values']=("Mã GV","Tên GV","Khoa")
combo_Search.grid(row=0, column=1, ipady=1.5, ipadx=1)
        
text_Search = Entry(Detail_Frame, width=30, bd=5, relief=GROOVE)
text_Search.grid(row=0, column=2, pady=10, padx=20, sticky="w")
        
Btn_Search = Button(Detail_Frame, text="Search", width=10, pady=5, command=search_info)
Btn_Search.grid(row=0, column=3, pady=10, padx=10)
Btn_ShowAll = Button(Detail_Frame, text="Chọn và Xoá", width=10, pady=5, command=delete_tree)
Btn_ShowAll.grid(row=0, column=4, pady=10, padx=10)
        
# Table Frame
Table_Frame = Frame(Detail_Frame, bd=4, relief=RIDGE, bg="white")
Table_Frame.place(x=10, y=70, width=820, height=460)
        
scroll_x = Scrollbar(Table_Frame, orient=HORIZONTAL)
scroll_y = Scrollbar(Table_Frame, orient=VERTICAL)
Teacher_Table = ttk.Treeview(Table_Frame, columns=("Mã GV", "Tên GV", "Giới Tính", "Khoa", "Số Điện Thoại", "Email", "Địa Chỉ"), xscrollcommand=scroll_x.set, yscrollcommand=scroll_y.set)
scroll_x.pack(side=BOTTOM, fill=X)
scroll_y.pack(side=RIGHT, fill=Y)
scroll_x.config(command=Teacher_Table.xview)
scroll_y.config(command=Teacher_Table.yview)
        
Teacher_Table.heading("Mã GV", text="Mã GV")
Teacher_Table.heading("Tên GV", text="Tên GV")
Teacher_Table.heading("Giới Tính", text="Giới Tính")
Teacher_Table.heading("Khoa", text="Khoa")
Teacher_Table.heading("Số Điện Thoại", text="Số Điện Thoại")
Teacher_Table.heading("Email", text="Email")
Teacher_Table.heading("Địa Chỉ", text="Địa Chỉ")
        
Teacher_Table['show'] = 'headings'
Teacher_Table.column("Mã GV", width=100)
Teacher_Table.column("Tên GV", width=100)
Teacher_Table.column("Giới Tính", width=100)
Teacher_Table.column("Khoa", width=100)
Teacher_Table.column("Số Điện Thoại", width=100)
Teacher_Table.column("Email", width=100)
Teacher_Table.column("Địa Chỉ", width=200)
        
Teacher_Table.pack(fill=BOTH, expand=1)

root.mainloop()
```
