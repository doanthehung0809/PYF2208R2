### Bài 7: Một công ty vận tải có cách tính chi phí giao hàng như sau:
1. 10kilogram đầu tiên: \$5/kg
2. Từ kilogram thứ 11 đến 20: \$7/kg
3. Từ kilogram thứ 20 trở lên: \$10/kg

Viết chương trình nhập vào số kg cần giao(số nguyên) và in ra chi phí giao hàng tương ứng


````python
kg = int(input("Nhập số lượng kg cần giao: "))
if kg<=0: print("Nhập lại dữ liệu kg phải >0")
if kg<=10: print(f"Phí giao {kg} kg là: {kg*5}$")
elif kg<=20: print(f"Phí giao {kg} kg là: {10*5+(20-kg)*7}$")
else: print(f"Phí giao {kg} kg là: {10*5+20*7+(kg-20)*10}$")
````

### Bài 10:
Tại một cửa hàng nọ có menu như sau:

1. Gà rán giá 35k
2. Hamburger giá 50k
3. Pesi giá 10k

Nhập số lượng món hàng cần mua

Tính và in ra tổng giá tiền

Nếu giá trị đơn hàng trên 2 triệu, giảm 10% cho khách, trên 4 triệu, giảm 15%

Tính và in ra tổng giá tiền được giảm

Tính và in ra tổng giá tiền khách cần trả

In ra dòng chữ "Cảm ơn quý khách đã mua hàng"

````python
"""Ví dụ:
Nhập số lượng gà rán khách mua:100
Nhập số lượng hamburger khách mua:100
Nhập số lượng pesi khách mua:15
Tổng giá tiền đơn hàng là: 8.650.000
Tổng tiền được giảm là: 1.297.500
Tổng tiền phải thanh toán: 7.352.500
Cảm ơn quý khách đã mua
"""
````

````python
Ga_ran = int(input("Nhập số lượng gà rán mua: "))
Hamburger = int(input("Nhập số lượng Hamburger mua: "))
Pesi = int(input("Nhập số lượng Pesi mua: "))
tong_bill = Ga_ran*35000 + Hamburger*50000 + Pesi*10000
if Ga_ran>=0 and Hamburger>=0 and Pesi>=0 and 0<tong_bill<=2000000:
    print(f"Tổng tiền phải thanh toán: {tong_bill}")
elif Ga_ran>=0 and Hamburger>=0 and Pesi>=0 and tong_bill>2000000:
    print(f"Tổng giá tiền được giảm là: {tong_bill*0.1}")
    print(f"Tổng tiền phải thanh toán: {tong_bill*(1-0.1)}")
elif Ga_ran>=0 and Hamburger>=0 and Pesi>=0 and tong_bill>4000000:
    print(f"Tổng giá tiền được giảm là: {tong_bill*0.15}")
    print(f"Tổng tiền phải thanh toán: {tong_bill*(1-0.15)}")
else: print("Nhập lại dữ liệu có giá trị âm")
````

### Bài 11:

Nhập vào chiều cao (cm) và cân nặng (kg), tính số BMI và xét rồi xin kết quả theo dữ liệu sau:

- BMI < 16: Gầy cấp độ III
- 16 <= BMI < 17:  Gầy cấp độ II
- 17<= BMI < 18.5: Gầy cấp độ I
- 18.5 <= BMI < 25: Bình thường
- 25 <= BMI < 30: Thừa cân
- 30 <= BMI < 35 : Béo phì cấp độ I
- 35 <= BMI < 40: Béo phì cấp độ II
- BMI > 40: Béo phì cấp độ III


````python
chieu_cao = float(input("Nhập chiều cao (cm): "))
can_nang = float(input("Nhập cân nặng (kg): "))
BMI = can_nang/(chieu_cao*2/100)
if BMI>40: print("Béo phì cấp độ III")
elif BMI>=30: print("Béo phì cấp độ II")
elif BMI>=25: print("Thừa cân")
elif BMI>=18.5: print("Bình thường")
elif BMI>=17: print("Gầy cấp độ I")
elif BMI>=16: print("Gầy cấp độ II")
else: print("Gầy cấp độ III")
````
