```python
a = float(input("Nhập điểm TB bạn a "))
b = float(input("Nhập điểm TB bạn b "))
c = float(input("Nhập điểm TB bạn c "))
if a>b and a>c: print("điểm của a lớn nhất")
elif b>a and b>c: print("điểm của b lớn nhất")
else: print("điểm của c lớn nhất")
```

## Cấu trúc lặp while

### Ví dụ: In ra tất cả số chẵn từ 0 đến n


```python
n =int(input("input n: "))
count = 0
while count<=n:
    print(count)
    count+=2
```

### Ví dụ: In ra tất cả số lẻ từ n về 1


```python
n = int(input("input n: "))
while n>=1:
    if n % 2 == 1:
        print(n)
    n-=1
```

### Ví dụ: Đảo ngược chuỗi

```python
string = input("input string: ")
index = len(string)-1
while index>=0:
    print(string[index],end="")
    index-=1
```

### Ví dụ: 
Nhập vào một số nguyên a

Cho S(k) = 1 + 2 + 3 + ... + k

Lựa chọn k sao cho S(k) > a với k nhỏ nhất

In S(k) và k lên màn hình

```python
a = int(input("Nhập vào số nguyên: "))
S = 0
k = 1
while S<=a:
    S+=k
    k+=1
print(S)
print(k-1)
```

### Lệnh Break

Cú pháp:

`break`

Lệnh break sẽ giúp hủy vòng lặp đang thực hiện

##### Lưu ý:

##### - Lệnh break <b>BẮT BUỘC</b> phải được đặt trong vòng lặp
##### - Lệnh break sẽ hủy vòng lặp gần nhất (trong trường hợp có nhiều vòng lắp lồng vào nhau)

#### Ví dụ

Sử dụng lệnh `while (True)` và break và in ra 10 lần câu "Hello mọi người"

```python
while True:
    print("hello world")
    break
```

## Lệnh continue

```python
# in ra những số chia hết cho 4 trong khoảng 0 -20 trừ số 8
```

```python
count = 0
while count <= 20:
    count += 4 
    if count ==8:
        continue
    print(count)
```
