```python
import turtle
import math
t=turtle.Turtle()
t.speed(40)
canh_mai=100
canh_than=150
#Mái trước tam giác
t.fillcolor('pink')
t.begin_fill()
t.left(135)
t.forward(canh_mai)
t.left(90)
t.forward(canh_mai)
t.left(135)
t.goto(0,0)
t.end_fill()
#Mái vuông phía sau
t.fillcolor('orange')
t.begin_fill()
t.left(45)
t.forward(canh_mai)
t.left(90)
t.forward(canh_mai)
t.left(90)
t.forward(canh_mai)
t.left(90)
t.forward(canh_mai)
t.end_fill()
#Phía trước nhà
t.fillcolor('blue')
t.begin_fill()
t.right(45)
t.forward(canh_than)
t.right(90)
t.forward(canh_mai*math.sqrt(2))
t.right(90)
t.forward(canh_than)
t.goto(0,0)
t.end_fill()
#Phía mặt sau
t.fillcolor('yellow')
t.begin_fill()
t.right(45)
t.forward(canh_mai)
t.right(135)
t.forward(canh_than)
t.right(45)
t.forward(canh_mai)
t.end_fill()
#Cửa chính phía trước
t.fillcolor('green')
t.begin_fill()
t.goto(-canh_mai*math.sqrt(2)/3,-canh_than)
t.goto(-canh_mai*math.sqrt(2)/3,-canh_than/3)
t.goto(-canh_mai*math.sqrt(2)*2/3,-canh_than/3)
t.goto(-canh_mai*math.sqrt(2)*2/3,-canh_than)
t.end_fill()
#Cửa sổ phía sau
t.penup()
t.fillcolor('brown')
t.begin_fill()
t.goto(canh_mai/5,-canh_than/5)
t.left(45)
t.pendown()
t.forward(canh_than/5)
t.left(135)
t.forward(canh_than/5)
t.left(45)
t.forward(canh_than/5)
t.left(135)
t.forward(canh_than/5)
t.end_fill()
#Thân Cây
t.penup()
t.goto(canh_mai*2,-canh_than)
t.left(135)
t.fillcolor('brown')
t.begin_fill()
t.forward(canh_mai/4)
t.left(90)
t.forward(canh_than/3)
t.left(90)
t.forward(canh_mai/4)
t.left(90)
t.forward(canh_than/3)
t.end_fill()
#Tán 1
t.penup()
t.right(180)
t.forward(canh_than/3)
t.left(90)
t.fillcolor('green')
t.begin_fill()
t.pendown()
t.forward(canh_mai/5)
t.right(120)
t.forward(canh_mai/4+2*canh_mai/5)
t.right(120)
t.forward(canh_mai/4+2*canh_mai/5)
t.right(120)
t.forward(canh_mai/4+2*canh_mai/5)
t.end_fill()
#Tán 2
t.penup()
t.right(120)
t.forward(canh_mai/4+2*canh_mai/5)
t.left(120)
t.pendown()
t.fillcolor('green')
t.begin_fill()
t.forward((canh_mai/4+2*canh_mai/5)/2)
t.right(120)
t.forward(canh_mai/4+2*canh_mai/5)
t.right(120)
t.forward(canh_mai/4+2*canh_mai/5)
t.right(120)
t.forward(canh_mai/4+2*canh_mai/5)
t.end_fill()
#Tán 3
t.penup()
t.right(120)
t.forward(canh_mai/4+2*canh_mai/5)
t.left(120)
t.pendown()
t.fillcolor('green')
t.begin_fill()
t.forward((canh_mai/4+2*canh_mai/5)/2)
t.right(120)
t.forward(canh_mai/4+2*canh_mai/5)
t.right(120)
t.forward(canh_mai/4+2*canh_mai/5)
t.right(120)
t.forward(canh_mai/4+2*canh_mai/5)
t.end_fill()
#Mặt trời
t.penup()
t.goto(canh_mai,canh_than*1.5)
t.pendown()
t.fillcolor('red')
t.begin_fill()
t.circle(canh_mai/4)
t.end_fill()
#Tia dài 1
t.right(90)
t.forward(canh_than/5)
#Tia ngắn 1
t.penup()
t.right(180)
t.forward(canh_than/5+canh_mai/4)
t.left(135)
t.forward(canh_mai/4)
t.pendown()
t.forward(canh_mai/5)
#Tia dài 2
t.penup()
t.right(180)
t.forward(canh_mai/4+canh_mai/5)
t.left(135)
t.forward(canh_mai/4)
t.pendown()
t.forward(canh_than/5)
#Tia ngắn 2
t.penup()
t.right(180)
t.forward(canh_than/5+canh_mai/4)
t.left(135)
t.forward(canh_mai/4)
t.pendown()
t.forward(canh_mai/5)
#Tia dài 3
t.penup()
t.right(180)
t.forward(canh_mai/4+canh_mai/5)
t.left(135)
t.forward(canh_mai/4)
t.pendown()
t.forward(canh_than/5)
#Tia ngắn 3
t.penup()
t.right(180)
t.forward(canh_than/5+canh_mai/4)
t.left(135)
t.forward(canh_mai/4)
t.pendown()
t.forward(canh_mai/5)
#Tia dài 4
t.penup()
t.right(180)
t.forward(canh_mai/4+canh_mai/5)
t.left(135)
t.forward(canh_mai/4)
t.pendown()
t.forward(canh_than/5)
#Tia ngắn 4
t.penup()
t.right(180)
t.forward(canh_than/5+canh_mai/4)
t.left(135)
t.forward(canh_mai/4)
t.pendown()
t.forward(canh_mai/5)
turtle.done()
```
